```python
class Student():

  def __init__(self, name, grade):
    self.name = name
    self.grade = grade



class StudentManager():

  def __init__(self):
    self.listOfStudents = []

  def accept(self, Student):
    self.listOfStudents.append(Student)

  def display(self):

    for crt in self.listOfStudents:
      print(crt.name+' '+ str(crt.grade) + '\n')

  def search(self, st):
      for crt in self.listOfStudents:
        if crt.name == st:
          print(crt.name+' '+str(crt.grade) + '\n')
          return

      print(st + 'Not found')

  def delete(self, st):
     for crt in range(len(self.listOfStudents)):
        if self.listOfStudents[crt].name == st:
          self.listOfStudents.remove(self.listOfStudents[crt])
          self.display()
          return
     print(st + ' Not found')

# * **
# list of info about the student
# problema ca dau st

  def update(self, st, **new_info):

      info = [st.name, st.grade]

      for crt in range(len(self.listOfStudents)):
        for i in range(info):
            if self.listOfStudents[crt].name == st:
              if type(info[i]) == type(new_info[i]):
                info[i] = new_info[i]

        return

      print(st + ' Not found')


sm = StudentManager()


while 1:

  opt = int(input('Da optiunea: '))

  match opt:
    case 0:
      print('Ai ales sa iesi')
      break

    case 1:
      n1 = input('Ai ales accept, alege nume: ')
      g1 = int(input('Alege nota'))
      s1 = Student(n1, g1)
      sm.accept(s1)


    case 2:
      print('Ai ales display: ')
      sm.display()


    case 3:
      n3 = input('Ai ales search, alege studentul: ')
      sm.search(n3)

    case 4:
      n4 = input('Ai ales delete, alege studnetul: ')
      sm.delete(n4)



    case 5:

      n5 = input('Ai ales update, alege studentul: ') # TO DO update la ce camp doresc

      info = []

      nv = input('Alege si noul nume: ')
      ng = input('Alege noua nota: ')

      if nv != '.':
        info.append(nv)

      if ng != '.':
        ng = int(ng)
        info.append(ng)


      sm.update(n5, info)
```

https://api-ninjas.com/api/planets



```python
import requests

planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune', 'Pluto', 'Kepler-452b', 'Kepler-62f', 'Gliese 667 Cc', 'Gliese 581c', '51 Pegasi b', 'PA-99-N2', 'M51-ULS-1b']

final_planets = pd.DataFrame()

for crt in planets:

  name = crt
  api_url = 'https://api.api-ninjas.com/v1/planets?name={}'.format(name)
  response = requests.get(api_url, headers={'X-Api-Key': '80zMrjqYq/XyXq5gPELXLw==hbDmEUX8aELIbAMj'})
  if response.status_code == requests.codes.ok:


    data = json.loads(response.text)

    planets_df = pd.DataFrame(data)

    planets_df = pd.json_normalize(data)

    final_planets = pd.concat([final_planets, planets_df], ignore_index=True)




  else:
    print("Error:", response.status_code, response.text)


display(final_planets)
```

```python
corr_df = final_planets

methods = ['pearson', 'kendall', 'spearman']

def find_corr(ser1, ser2, ser3, method):

  for m in methods:
    ser1.corr(ser2, method = m)
    ser2.corr(ser3, method = m)
    ser3.corr(ser1, method = m)

ser_light_years = corr_df['distance_light_year']
ser_period = corr_df['distance_light_year']
ser_temp = corr_df['temperature']

p = ser_light_years.corr(ser_period, method = 'kendall')
k = ser_temp.corr(ser_period, method = 'kendall')
s = ser_light_years.corr(ser_temp, method = 'kendall')

print(p)

print(k)

print(s)


```
